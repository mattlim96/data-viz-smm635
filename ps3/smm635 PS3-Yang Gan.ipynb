```python
# setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# set seed
np.random.seed(111)

# period 
p = 11

# No. of new ventures
s = 1000

# Assuming that the ventures' revenues show exponential growth every year
e = []
for i in range(s):
    for j in np.arange(1,p,1):
        r = np.exp(j) * np.random.uniform(0.1,1)
        e.append([i,j,r])
        
e = pd.DataFrame(e, columns = ['i','j', 'r'])
```

```python
# Calculate the mean of the 1000 ventures growth
c = pd.DataFrame(e.groupby('j', as_index = False)['r'].agg(np.mean))
```

```python
# create figure
fig = plt.figure(figsize = (9,6))

# add the chart
ax = fig.add_subplot(111)

# plot the data
x = np.arange(1,p,1)

for i in range(s):
    y_obs = e.loc[(e['i'] == i)]['r'].values
    ax.plot(x, y_obs, color = 'deepskyblue', alpha = 0.01)

y_e = c.r
ax.plot(x, y_e, marker = 'o', markersize = 10, alpha = 1)

# set title
ax.set_title('Individual and Average revenues')

# axes
ax.set_xlabel('Time')
ax.set_ylabel('Revenue')

# show plot
plt.show()
```
