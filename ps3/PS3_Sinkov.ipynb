```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(666)
# periods
p = 11
# sub-population size
s = 1000
data = []
for i in range(s):
    for j in np.arange(1, p, 1):
        r = np.exp(j) * np.random.uniform(0.1, 1)
        data.append([i, j, r,])
        
data = pd.DataFrame(data, columns=['Company', 'Year', 'Revenue',])
```

```python
fig, axs = plt.subplots(ncols=5, nrows=3,figsize=(20,12))
gs = axs[1, 2].get_gridspec()
for ax in axs[2,0:5]:
    ax.remove()
axbig = fig.add_subplot(gs[2,0:5])
axs = axs.ravel()
x = np.arange(1, 11, 1)
for l in range(1000):
    y = data.loc[data['Company'] == l]['Revenue'].values
    axbig.plot(x, y, color='lightcoral', alpha=0.03)
    axbig.set_xlabel('Years')
    axbig.set_ylabel('Revenue')
    axbig.set_xticks(range(1,11, 1))
    axbig.spines['right'].set_visible(False)
    axbig.spines['bottom'].set_visible(False)
    axbig.spines['left'].set_visible(False)
        
for i in range(10):
    data[data.Year==i+1].Revenue
    axs[i].hist(data[data.Year==i+1].Revenue,bins=100,color='lightcoral')
    axs[i].set_title('Year '+ str(i+1))
    axs[i].set_yticks(range(0, 25, 5))
    axs[i].axes.yaxis.set_visible(False)
    axs[i].spines['top'].set_visible(False)
    axs[i].spines['right'].set_visible(False)
    axs[i].spines['bottom'].set_visible(False)
    axs[i].spines['left'].set_visible(False)
    if i == 0 or i ==5:
        axs[i].axes.yaxis.set_visible(True)
        axs[i].set_ylabel('Number of Companies')
    if i == 7:
        axs[i].set_xlabel('Revenue')
```
