Use NumPy to simulate yearly revenues for 1,000 new ventures for a ten-year timespan (since the inception of businesses to year '10'). Then, use Matplotlib to create one figure that shows both the within and between variation in revenues.

# Solution 1

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s=1000
dict = {'2001':np.random.randint(low=5000, high=30000,size=s),
               '2002':np.random.randint(low=5000, high=30000,size=s),
               '2003':np.random.randint(low=5000, high=30000,size=s),
               '2004':np.random.randint(low=5000, high=30000,size=s),
               '2005':np.random.randint(low=5000, high=30000,size=s),
               '2006':np.random.randint(low=5000, high=30000,size=s),
               '2007':np.random.randint(low=5000, high=30000,size=s),
               '2008':np.random.randint(low=5000, high=30000,size=s),
               '2009':np.random.randint(low=5000, high=30000,size=s),
               '2010':np.random.randint(low=5000, high=30000,size=s)}
df_rev = pd.DataFrame(dict)
df_rev
```

```python
#creating a bopx plot to show the variation within and between 1000 companies over 10 yrs
df_rev.boxplot()
```

# Solution 2

```python
dfRevenue = df_rev.transpose()
dfRevenue
```

```python
dfRevenue['Min'] = dfRevenue.min(axis=1)
dfRevenue['Max'] = dfRevenue.max(axis=1)
dfRevenue['Mean'] = dfRevenue.mean(axis=1)
dfRevenue['Std_Dev'] = dfRevenue.std(axis=1)
dfRevenue['Variance'] = dfRevenue.var(axis=1)
dfRevenue
```

```python
#creating a seperate df with the statistics of the initial df

data = [dfRevenue['Min'],dfRevenue['Max'],dfRevenue['Mean'],dfRevenue['Std_Dev'],dfRevenue['Variance']]
df_R = pd.concat(data, axis=1)
df_R
```

```python
##Creating a side-by-side bar graph to show the mean, std dev (variations) withint and between over the 10yr span

# Numbers of pairs of bars you want
N = 10

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = df_R['Mean']
# Specify the values of orange bars (height)
orange_bar = df_R['Std_Dev']


# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(10,10))

# Width of a bar 
width = 0.4      

# Plotting
plt.bar(ind, blue_bar , width, label='Mean')
plt.bar(ind + width, orange_bar, width, label='Std_Dev')

plt.xlabel('Years')
plt.ylabel('Revenue ($)')
plt.title('Variation in Revenue over 10 years')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, ('2001', '2002', '2003','2004','2005','2006','2007','2008','2009','2010'))

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```

# Solution 3

```python
## Creating the data set

# set seed
np.random.seed(666)
p = 10
s = 1000

#simulating random revenues for 1000 new ventures over 10-year span
revenue_2001 = np.random.randint(low=5000, high=30000,size=s)
revenue_2002 = np.random.randint(low=5000, high=30000,size=s)
revenue_2003 = np.random.randint(low=5000, high=30000,size=s)
revenue_2004 = np.random.randint(low=5000, high=30000,size=s)
revenue_2005 = np.random.randint(low=5000, high=30000,size=s)
revenue_2006 = np.random.randint(low=5000, high=30000,size=s)
revenue_2007 = np.random.randint(low=5000, high=30000,size=s)
revenue_2008 = np.random.randint(low=5000, high=30000,size=s)
revenue_2009 = np.random.randint(low=5000, high=30000,size=s)
revenue_2010 = np.random.randint(low=5000, high=30000,size=s)
```

```python
#mean of each year's revenue (height of bars in bar plot)

mean_2001 = np.mean(revenue_2001)
mean_2002 = np.mean(revenue_2002)
mean_2003 = np.mean(revenue_2003)
mean_2004 = np.mean(revenue_2004)
mean_2005 = np.mean(revenue_2005)
mean_2006 = np.mean(revenue_2006)
mean_2007 = np.mean(revenue_2007)
mean_2008 = np.mean(revenue_2008)
mean_2009 = np.mean(revenue_2009)
mean_2010 = np.mean(revenue_2010)

#standard deviation of each year's revenue (height of error bars)

std_2001 = np.std(revenue_2001)
std_2002 = np.std(revenue_2002)
std_2003 = np.std(revenue_2003)
std_2004 = np.std(revenue_2004)
std_2005 = np.std(revenue_2005)
std_2006 = np.std(revenue_2006)
std_2007 = np.std(revenue_2007)
std_2008 = np.std(revenue_2008)
std_2009 = np.std(revenue_2009)
std_2010 = np.std(revenue_2010)
```

```python
#defining labes, positions, box plot bar heigths, and error bar heights

labels = ['2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']
x_pos = np.arange(len(labels))
mean = [mean_2001,mean_2002,mean_2003,mean_2004,mean_2005,mean_2006,mean_2007,mean_2008,mean_2009,mean_2010]
error = [std_2001,std_2002,std_2003,std_2004,std_2005,std_2006,std_2007,std_2008,std_2009,std_2010]

#building the plot
fig, ax = plt.subplots()
ax.bar(x_pos, mean, yerr=error, align='center',alpha=0.5, ecolor='black',capsize=10)

#axis label
ax.set_xlabel('Time since the inception of business (years)',fontsize=12)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_ylabel('Avg. Yearly Revenue ($)',fontsize=12)
ax.set_title('Variations in Revenue for 1000 businesses')

#axis layout
ax.yaxis.grid(True)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#plot
plt.show()
```
