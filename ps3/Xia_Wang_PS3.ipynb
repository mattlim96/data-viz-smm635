```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(111)

#periods
p = 11

#size
s = 1000
```

```python
#exponential 
e = []

for i in range(s):
    for j in np.arange(1, p, 1):
        r = np.exp(j) * np.random.uniform(0.4, 1)
        e.append([i, j, r, 'e'])
        
        
e = pd.DataFrame(e, columns=['i', 'j', 'r', 'g'])


        
u = pd.DataFrame(u, columns=['i', 'j', 'r', 'g'])

# log shape
b = 5000

l = []

for i in range(s):
    for j in np.arange(1, p, 1):
        r = b * np.log(j) * np.random.uniform(0.1, 1)
        l.append([i, j, r, 'l'])


l = pd.DataFrame(l, columns=['i', 'j', 'r', 'g'])

# append data
df = pd.concat([u, l], axis=0, )
```

Within and between approach

```python
# create figure
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

# plot data
x = np.arange(1, 11, 1)

# plot data
x = np.arange(1, 11, 1)

y_e = [0.55 * np.exp(i) for i in np.arange(1, 11, 1)]
ax.plot(x, y_e, color='cornflowerblue', alpha=1, marker='o', markersize=10, label='Exponential growth')
    
y_l = [0.55 * (b * np.log(x)) for x in np.arange(1, 11, 1)]
ax.plot(x, y_l, color='orange', alpha=1, marker='o', markersize=10, label='Plateau')


# legend
ax.legend(loc='best')

# title
ax.set_title('Annual revenue trajectories')

# axes
ax.set_xlabel('Time elapsed since incorporation (years)')
ax.set_ylabel('Annual revenue (K)')
ax.grid(axis='both',linestyle='--', linewidth=0.2)

ax.fill_between(x,y_e,y_l, where=(x<10),interpolate=True ,alpha=0.2, color='orange')
ax.fill_between(x,y_e,y_l, where=(x>=10),interpolate=True ,alpha=0.2, color='blue')

ax.text(4.5, 2500, "Revenue gap", fontsize=16, color='grey')
ax.text(2.3, 6000, "Plateau > Exponential", fontsize=10, color='grey')
ax.text(7, 10000, "Exponential > Plateau", fontsize=10, color='grey')
ax.text(9.6, 6600, "Revenue gap", fontsize=14, color='grey',rotation=82)
ax.arrow(3.5,5500,0.8, -2700, head_width=.1,color='grey' )
ax.arrow(8.2,9700,1.45, -1000, head_width=0.1, color='grey')
        
# show plot
plt.show()
```
