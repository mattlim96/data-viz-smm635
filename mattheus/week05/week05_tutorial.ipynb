```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc
```

```python
categories = ['A','B','C','D']
counts = [12,14,8,6]
pos = np.linspace(0,len(categories)-1,len(categories))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(pos,counts,width=0.5)
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.bar(pos,counts,width=0.01,color='k',alpha=0.5)

ax.set_xticks(pos)
ax.set_xticklabels(categories)
ax.set_yticks(np.arange(0,max(counts),4))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
# ax.axis('off')

plt.show()
```

```python
mu, sigma, = 2.5, 4
bins = 50
x = np.random.normal(mu, sigma, 1000)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

n, bins, patches = ax.hist(x, bins, density=True, histtype='stepfilled', alpha=0.25,
                           cumulative=True, label='Empirical', color='m')

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
y = y.cumsum()
y /= y[-1]

ax.plot(bins, y, 'k--', linewidth=1.5, label='Theoretical', color='k')

ax.hist(x, bins, density=True, histtype='stepfilled', cumulative=-1, alpha=0.25,
        label='Reversed emp.', color='c')

ax.grid(ls='--', color='k', alpha=0.5)

ax.legend(loc='best')
```
