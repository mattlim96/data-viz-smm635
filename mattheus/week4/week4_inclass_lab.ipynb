```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

def get_forward_month_list():
    now = datetime.strptime('2020-01-01', '%Y-%m-%d')
    return [(now + relativedelta(months=i)).strftime('%b') for i in range(12)]

time = get_forward_month_list()
```

```python
np.random.seed(0)
numCures = np.random.randint(1, 100, size=12)
# time = np.arange(1,13,1)
# df = pd.DataFrame(zip(time,numCures),columns=['Months','# of Cures'])
```

```python
plt.plot(time,numCures)
plt.title('Time Series of # of Cures')
plt.show()
```
